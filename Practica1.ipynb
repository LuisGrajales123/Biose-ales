```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = np.array([3.1,1, -0.5, -3.2, 6])
b= np.array([1, 3, 2.2, 5.1, 1])

producto_escalar = np.dot(a,b)

print("El producto escalar es:",producto_escalar)

punto_punto = a*b

print("El producto punto a punto entre los vectores es:",punto_punto)
```

La función np.dot() de NumPy se usa para calcular el producto punto (producto escalar) entre arreglos.

```python
A = np.array([[2,-1,3],[4, 1.5 ,-2.5 ],[7.3, -0.9, 0.2]])

AT= A.T
print("La matriz transpuesta es: ", AT)
```

Función de comandos de NumPy:

Ones(): Crea una matriz llena de unos.
round():Redondea números al número de decimales que indiques.
ceil():Redondea siempre hacia arriba.
floor():Redondea siempre hacia abajo.

```python
M = np.ones((3, 4))
print("La función ones hace: ",M)

x = np.array([3.1416, 2.7182, 1.5555])
print("La función round hace: ",np.round(x, 2))

x = np.array([2.1, 3.7, 4.2])
print("La función cell hace: ",np.ceil(x))

x = np.array([2.9, 3.7, 4.2])
print("La función onesfloor: ",np.floor(x))
```

```python
#Señales 
n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)
y2 = np.cos(2 * np.pi * 0.03 * n)
s = y + y2
t = y * y2

# Grafica 1
plt.figure()
plt.plot(n, y, label="y[n] = sin(pi*0.12n)")
plt.plot(n, y2, label="y2[n] = cos(2pi*0.03n)")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.title("Señales y[n] y y2[n]")
plt.legend()
plt.grid()

# Grafica 2
plt.figure()
plt.plot(n, s, label="s[n] = y[n] + y2[n]")
plt.plot(n, t, label="t[n] = y[n] * y2[n]")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.title("Señales suma y producto")
plt.legend()
plt.grid()

plt.show()
```

```python
def estadisticas_notas(dic_notas):
    # convertir a Serie
    notas = pd.Series(dic_notas)
    
    # calcular estadísticas
    resultado = pd.Series({
        'Minima': notas.min(),
        'Maxima': notas.max(),
        'Media': notas.mean(),
        'Desviacion_tipica': notas.std()
    })
    
    return resultado

def agregar_BMI(df):
    # convertir altura de cm a metros
    altura_m = df['Height'] / 100
    
    df['BMI'] = df['Weight'] / (altura_m ** 2)
    
    return df
```

```python
df = pd.read_csv("datos.csv", sep=";")
print(df.head())

df = df.drop(columns=["Unnamed: 0"])

df = agregar_BMI(df)
print(df.head())

def clasificar_bmi(df):
    
    def categoria(valor):
        if valor < 18.5:
            return 'Bajo peso'
        elif valor < 25:
            return 'Normal'
        elif valor < 30:
            return 'Sobrepeso'
        else:
            return 'Obesidad'
    
    df['Categoria_IMC'] = df['BMI'].apply(categoria)
    return df

df = clasificar_bmi(df)
print(df[['Weight','Height','BMI','Categoria_IMC']].head())
```
