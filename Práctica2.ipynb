```python
import numpy as np
import matplotlib.pyplot as plt

def impseq(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = ((n-n0) == 0).astype(int)
    return x, n

# ejemplo: impulso en n0=2 entre -5 y 5
x, n = impseq(2, -5, 5)

plt.stem(n, x)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Secuencia Impulso")
plt.grid(True)
plt.show()
```

```python
def unitario(n0, n1, n2):
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n - n0) >= 0             # condicion para que sea 1 o True
 x = x.astype(int)             # convertir True/False -> 1/0
 return [x,n]

x, n = unitario(0, -5, 5)
plt.stem(n, x)
plt.title("Escalón unitario u(n)")
plt.xlabel("n")
plt.ylabel("u(n)")
plt.grid(True)
```

```python
def rampa(n0, n1, n2):
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n - n0) * (n >= n0)      # condicion para que sea 1 o True
 return [x,n]

#Ejemplo
x, n = rampa(0.2, -4, 4)
plt.stem(n, x)
plt.title("Rampa r(n)")
plt.xlabel("n")
plt.ylabel("r(n)")
plt.grid(True)
plt.show()
```

```python

# Escalar real
print(np.exp(1))   # e^1 = 2.718...

# Vector
print(np.exp([0, 1, 2]))   # [1.0, e, e^2]

# Número complejo
print(np.exp(1 + 1j))   # e^(1+j) = e^1 * e^(j) = 2.718*(cos1 + j sin1)

# Parámetros
sigma = -0.3   # parte real (define atenuación si es < 0)
w0 = np.pi/4   # frecuencia angular
n = np.arange(0, 30)  # vector de tiempo discreto

# Señal exponencial compleja
x = np.exp((sigma + 1j*w0) * n) #se esta elevando a e la funcion compleja

# Graficar parte real e imaginaria
plt.stem(n, np.real(x), linefmt='b-', label='Real')
plt.stem(n, np.imag(x), linefmt='r-', label='Imaginario')
plt.title("Exponencial compleja")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.legend()
plt.show()
```

```python
def impseq(n0, n1, n2):
# Genera x(n) = delta(n-n0); n1 <= n <= n2
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 return [x,n]

n1 = -4
n2 = 1

# Crear los impulsos con las amplitudes correspondientes
imp1, n = impseq(-1, n1, n2)  # 3δ(n + 1)
imp2, _ = impseq(-3, n1, n2)  # 5δ(n + 3)
imp3, _ = impseq(-2, n1, n2)  # 3δ(n + 2)
imp4, _ = impseq(0, n1, n2)   # 3δ(n) + δ(n)

# Crear la secuencia x1(n)
x1 = 3 * imp1 + 5 * imp2 + 3 * imp3 + 4 * imp4  # Sumar impulsos ponderados

# Graficar la función compuesta por impulsos
plt.figure(figsize=(8, 5))
plt.stem(n, x1)
plt.title('Función compuesta de impulsos x1(n) 3a')
plt.xlabel('n')
plt.ylabel('x1(n)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

```python
import numpy as np
# definir el vector de muestras
# Rango de muestras
n1 = -6
n2 = 7

# Impulsos
x1u, n = unitario(-4, n1, n2) #-2*u(n+4)
x2u, _ = unitario(-2, n1, n2) #3u(n+2)

#definir la ecuacion general
x2 = -2*x1u+3*x2u

import matplotlib.pyplot as plt

plt.stem(n, x2)
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.title('Secuencia x2[n] 3b')
plt.grid(True)
plt.show()
```

```python
def expo(n, n1, n2):
  e = np.exp(0.01 * n)
  return e

def seno(n, n1, n2):
  s = np.sin(2 * np.pi *n )
  return s

n=np.arange(0,101)
n1 = 0
n2 = n[-1]

exponencial = expo(n, n1, n2)
sinusoidal = seno(n,n1,n2)

secuencia = exponencial * sinusoidal

plt.figure(figsize=(8, 5))
plt.stem(n,secuencia,'b')
plt.title('Secuencia 3c')
plt.xlabel('Rango')
plt.ylabel('Resultado')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid()
plt.show()
```

```python
def unitario(n):
    u = np.where(n >= 0, 1, 0)
    return u

def rampa(n):
    r = np.maximum(n, 0)
    return r

n = np.arange(-10, 11)

x4 = 2 * rampa(n + 3) - rampa(n - 2) - 5 * unitario(n - 3)#se pasa la secuencia mostrada en la guía

plt.figure(figsize=(8, 5))
plt.stem(n, x4,'b')
plt.title('Secuencia 3d')
plt.xlabel('Rango')
plt.ylabel('Resultado')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid()
plt.show()
```

```python
# Generar la secuencia en forma de vector con 10 posiciones desde el 0 hasta el 17

#En la parte de arriba se creo la funcion de la rampa
#Se le aplica esa funcion para cada uno de los tramos
x1, n1 = rampa(0, 0, 5)     #n0 = 0, n1=0, n2=5 --> n1 es desde donde empieza y n2 hasta donde va el vector
x2, n2 = rampa(5, 6, 11)    #n0 = 5, n1 = 6, n2 = 11
x3, n3 = rampa(10, 12, 17)  #n0=10, n1=12, n2=17

# Concatenar todo en un solo vector
x_total = np.concatenate((x1, x2, x3))
n_total = np.concatenate((n1, n2, n3))

print("n_total (Posiciones) =", n_total)
print("x_total (Vector con valores) =", x_total)

#la derivada se define como una diferencia hacia adelante (n-1)
derivada = np.diff(x_total)
nd = np.arange(0, len(derivada)) #como se realizó la derivada, se eliminó una posición
print(derivada)

plt.stem(n_total, x_total, linefmt='b-', label='Original')
plt.stem(nd, derivada, linefmt='r-', label='Derivada')
plt.title("Secuencias obtenidas punto 4")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.legend()
plt.show()
```
